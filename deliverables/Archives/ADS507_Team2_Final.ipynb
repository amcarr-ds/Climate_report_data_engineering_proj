```python
#Transformation step - this is for country, year, total, temperature, and population only

Transform1 = pd.read_sql("""
SELECT e.Country, e.Year, e.Total, t.Value AS Temperature, 
wp.`2020 Population` AS Population
FROM emissions e
JOIN temperature t ON e.Country = t.Area AND e.Year = t.Year
JOIN worldpopulation wp ON e.Country = wp.`Country/Territory`
ORDER BY e.Country, e.Year""", conn)

Transform1
```
