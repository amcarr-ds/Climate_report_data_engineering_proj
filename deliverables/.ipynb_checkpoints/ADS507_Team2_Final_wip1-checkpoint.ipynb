```python
query01_stmnt = """
SELECT
    pt.country,
    pt.year,
    pt.population,
    eg.total_avg_1970,
    eg.coal_avg_1970
FROM population_trans AS pt
INNER JOIN (
SELECT
    country,
    record_year,
    avg(total) AS total_avg_1970,
    avg(coal) AS coal_avg_1970
FROM emissions_gross
WHERE record_year = 1970
GROUP BY country, record_year
) AS eg
    ON pt.country = eg.country AND pt.year = eg.record_year
"""
query01_rd = pd.read_sql(query01_stmnt, conn)

query01_rd
```

```python
query02a_stmnt = """
SELECT
    *,
    avg(eg.total) OVER (PARTITION BY pt.country) AS country_period_total_emission_avg,
    sum(tm.temp_increase) OVER (PARTITION BY pt.country) AS country_period_temp_inc
FROM population_trans AS pt
INNER JOIN (
    SELECT
        country,
        record_year,
        total
    FROM emissions_gross
    WHERE record_year = 1970
    GROUP BY country, record_year
) AS eg
    ON pt.country = eg.country
INNER JOIN (
    SELECT
        country,
        record_year,
        temp AS temp_increase
    FROM temperature
    WHERE record_year = 1970
    GROUP BY country, record_year
) AS tm
    ON pt.country = tm.country AND eg.record_year = tm.record_year
"""
query02a_rd = pd.read_sql(query02a_stmnt, conn)

query02a_rd
```

```python
query02b_stmnt = """
SELECT
    *,
    avg(eg.total) OVER (PARTITION BY pt.country) AS country_period_total_emission_avg,
    sum(tm.temp_increase) OVER (PARTITION BY pt.country) AS country_period_temp_inc
FROM population_trans AS pt
INNER JOIN (
    SELECT
        country,
        record_year,
        total
    FROM emissions_gross
    WHERE record_year = 1970
    GROUP BY country, record_year
) AS eg
    ON pt.country = eg.country
INNER JOIN (
    SELECT
        country,
        record_year,
        temp AS temp_increase
    FROM temperature
    WHERE record_year = 1970
    GROUP BY country, record_year
) AS tm
    ON pt.country = tm.country AND eg.record_year = tm.record_year
WHERE pt.year = 1970
UNION
SELECT
    *,
    avg(eg.total) OVER (PARTITION BY pt.country) AS country_period_total_emission_avg,
    sum(tm.temp_increase) OVER (PARTITION BY pt.country) AS country_period_temp_inc
FROM population_trans AS pt
INNER JOIN (
    SELECT
        country,
        record_year,
        total
    FROM emissions_gross
    WHERE record_year BETWEEN 1971 AND 1980
    GROUP BY country, record_year
) AS eg
    ON pt.country = eg.country
INNER JOIN (
    SELECT
        country,
        record_year,
        temp AS temp_increase
    FROM temperature
    WHERE record_year BETWEEN 1971 AND 1980
    GROUP BY country, record_year
) AS tm
    ON pt.country = tm.country AND eg.record_year = tm.record_year
WHERE pt.year = 1980
"""
query02b_rd = pd.read_sql(query02b_stmnt, conn)

query02b_rd
```
